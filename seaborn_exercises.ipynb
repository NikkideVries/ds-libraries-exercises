Question 4: Load the chipotle dataset from SQL, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
import env
```

```python
url = env.get_db_url('chipotle')
```

```python
pd.read_sql('show tables', url)
```

```python
orders = pd.read_sql('select * from orders', url)
```

```python
orders
```

```python
# clean up the data: 
def remove_issues(price):
    #remove commas from the string
    no_spaces = price.replace(',','')
    #remove dollar signs
    no_sign = no_spaces.replace('$','')
    return no_sign   
```

```python
orders['item_price'] = orders['item_price'].apply(remove_issues).astype(float)
```

```python
# another way to clean the data: 
# orders.item_price = orders.item_price.str.strip('$').astype(float)
```

```python
orders
```

```python
# groupby
popular_orders = orders.groupby('item_name').quantity.sum().nlargest(4)
popular_orders
```

```python
total_sum = orders.groupby('item_name').item_price.sum()
total_sum.head()
```

```python
popular_orders_and_price = pd.merge(popular_orders, total_sum, on = 'item_name', how = 'inner')
popular_orders_and_price
```

```python
# need to reset the index to not me the item_name:
popular_orders_and_price = popular_ordesr_and_price.reset_index()
```

```python
popular_orders_and_price
```

```python
# create the graph:
sns.barplot(data = popular_orders_and_price, y = 'item_name', x = 'item_price')

plt.title("Most popular Chipotle Items")
plt.xlabel('Total Revenue')
plt.show()
```
