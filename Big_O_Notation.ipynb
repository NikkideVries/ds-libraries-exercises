```python
#visualization imports
import matplotlib.pyplot as plt  

#standard DS imports
import numpy as np
import pandas as pd

#numerical imports
import math
from random import randint
import random
```

```python


plt.figure(figsize=(10,8))

# y = 0n + 1
x1 = np.linspace(-30,30,100)
y1 = [(0*n) + 1 for n in x1]
plt.plot(x1,y1, label = "O(1)", color = 'red')

# y = log(n)
x2 = range(1,31)
y2 = [np.log(x) for x in x2]
plt.plot(x2,y2, label = 'O(log n)', color ='orange')

# y = n
x3 = np.linspace(-30,30,100)
y3 = [n for n in x3]
plt.plot(x3,y3, label = 'O(n)', color = "yellow")

#y = n * log(n)
x4 = np.arange(1,31)
y4 = [n * (np.log(n)) for n in x4]
plt.plot(x4,y4, label = 'O(n log n)', color = "greenyellow")

#y = n^2
x5 = np.linspace(-30,30,100)
y5 = [n**2 for n in x5]
plt.plot(x5,y5, label = 'O($n^2$)', color = 'green')

#y = 2**n
x6 = np.linspace(-20,10,40)
y6 = [2**n for n in x6]
plt.plot(x6,y6, label = 'O($2^n$)', color = 'dodgerblue')

#y = n!
x7 = np.arange(0,7)
y7 = [math.factorial(n) for n in x7]
plt.plot(x7,y7, label = 'O(n!)', color = 'mediumpurple')

#y = n^n
x8 = np.linspace(0,4,20)
y8 = [n**n for n in x8]
plt.plot(x8,y8, label = 'O($n^n$)', color = 'pink')


plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
plt.legend(loc= 'upper center')
plt.grid(True, linewidth = 0.1, c= 'gray' )
plt.show()
```
