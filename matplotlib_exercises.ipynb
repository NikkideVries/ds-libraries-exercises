### Question 1: 
Use matplotlib to plot the following equation:
                            
  $$ y=x^2+2 $$

You'll need to write the code that generates the x and y points.
Add an anotation for the point 0, 0, as the origin.   



```python
plt.style.use('bmh')

x = range(-50, 51)
y = [n**2 - n + 2 for n in x]

plt.plot(x, y, color ="#33ff8e")
plt.title('$y=x^2-x+2$')
plt.xlabel('$x$')
plt.ylabel('$f(x)')
plt.annotate('Origin', xy=(0,0), xytext=(-4,300),\
             arrowprops=dict(color = 'k', arrowstyle="->"))
plt.show()
```

### Question 2:
Create and label 4 separate charts with the following equations (choose a range for x that makes sense):
- $y=\sqrt{x}$
- $y=x^3$
- $y=2^x$
- $y=1/(x+1)$


You can use functions from the `math` module to help implement some of the equations above.


$y=\sqrt{x}$

```python
x = np.linspace(0, 10, 100) #range of x values
y = np.sqrt(x)

plt.figure(figsize=(8,8))
plt.plot(x,y, label='$y=\sqrt{x}$', color = 'pink')
plt.title('Graph of $y=\sqrt{x}$')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10, 10, 100) #range of x values
y1 = x**3

plt.figure(figsize=(8,8))
plt.plot(x,y1, label='$y=x^3$', color = 'green')
plt.title('Graph of $y=x^3$')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10, 10, 100) #range of x values
y2 = 2**x

plt.figure(figsize=(8,8))
plt.plot(x,y2, label= '$y=2^x$', color = 'navy')
plt.title('Graph of $y=2^x$')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-10, 10, 100) #range of x values
y2 = 1/(x+1)

plt.figure(figsize=(8,8))
plt.plot(x,y2, label= '$y=1/(x+1)$', color = 'rebeccapurple')
plt.title('Graph of $y=1/(x+1)$')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

### Question 3: 
Combine the figures you created in the last step into one large figure with 4 subplots.

```python
x1 = np.linspace(0,5,100)
x2 = np.linspace(-10,10,100)
x3 = np.linspace(-5,5,100)
x4 = np.linspace(-5,5,100)


y1 = [np.sqrt(x) for x in x1]
y2 = [x**3 for x in x2]
y3 = [2**x for x in x3]
y4 = [1/(x+1) for x in x4]
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,8))

axs[0,0].plot(x1,y1, color='g')
axs[0,0].set_title('$y=\sqrt{x}$')

axs[0,1].plot(x2,y2, color='b')
axs[0,1].set_title('$y=x^3$')

axs[1,0].plot(x3, y3, color='purple')
axs[1,0].set_title('$y=2^x$')

axs[1,1].plot(x4,y4, color='limegreen')
axs[1,1].set_title('$y=1/(x+1)$')


plt.show()
```

```python

plt.figure(figsize=(10,8))
plt.suptitle('Different Equations')

plt.subplot(221)
plt.plot(x1, y1, color='b', ls='--')
plt.title('$y=\sqrt{x}$')

plt.subplot(222)
plt.plot(x2, y2, color="g", ls='--')
plt.ylim(-10,10)
plt.xlim(-10,10)
plt.title('$y=x^3$')

plt.subplot(223)
plt.plot(x3, y3, color='purple', ls='--')
plt.title('$y=2^x$')

plt.subplot(224)
plt.plot(x4, y4, color='limegreen', ls='--')
plt.title('$y=1/(x+1)$')

plt.show()
```
