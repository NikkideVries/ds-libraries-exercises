### Question 1: 
Use matplotlib to plot the following equation:
                            
  $$ y=x^2+2 $$

You'll need to write the code that generates the x and y points.
Add an anotation for the point 0, 0, as the origin.   



```python
plt.style.use('bmh')

x = range(-50, 51)
y = [n**2 - n + 2 for n in x]

plt.plot(x, y, color ="#33ff8e")
plt.title('$y=x^2-x+2$')
plt.xlabel('$x$')
plt.ylabel('$f(x)')
plt.annotate('Origin', xy=(0,0), xytext=(-4,300),\
             arrowprops=dict(color = 'k', arrowstyle="->"))
plt.show()
```

### Question 2:
Create and label 4 separate charts with the following equations (choose a range for x that makes sense):

- $y=x^3$

You can use functions from the `math` module to help implement some of the equations above.


$y=\sqrt{x}$

```python
x = np.linspace(1, 10, 100) #range of x values
y = np.sqrt(x)

plt.figure(figsize=(8,8))
plt.plot(x,y, label='$y=\sqrt{x}$', color = 'pink')
plt.title('Graph of $y=\sqrt{x}$')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(1, 10, 100) #range of x values
y1 = x**3

plt.figure(figsize=(8,8))
plt.plot(x,y1, label='$y=x^3$', color = 'green')
plt.title('Graph of $y=x^3$')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(1, 10, 100) #range of x values
y2 = 2**x

plt.figure(figsize=(8,8))
plt.plot(x,y2, label= '$y=2^x$', color = 'navy')
plt.title('Graph of $y=2^x$')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(1, 10, 100) #range of x values
y2 = 1/(x+1)

plt.figure(figsize=(8,8))
plt.plot(x,y2, label= '$y=1/(x+1)$', color = 'rebeccapurple')
plt.title('Graph of $y=1/(x+1)$')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

### Question 3: 
Combine the figures you created in the last step into one large figure with 4 subplots.

```python
 x = np.linspace(0, 5, 100)

y1 = np.sqrt(x)
y2 = x**3
y3 = 2**x
y4 = 1/(x+1)

plt.figure(figsize=(8,8))
plt.suptitle('Different Equations')

plt.subplot(221)
plt.plot(x, y1, color='b', ls='--')
plt.title('$y=\sqrt{x}$')

plt.subplot(222)
plt.plot(x, y2, color="g", ls='--')
plt.title('$y=x^3$')

plt.subplot(223)
plt.plot(x, y3, color='purple', ls='--')
plt.title('$y=2^x$')

plt.subplot(224)
plt.plot(x, y4, color='limegreen', ls='--')
plt.title('$y=1/(x+1)$')

plt.show()
```

### Question 4:
Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
# np.linspace(start, stop, num=50(number of samples))
x = np.linspace(-5, 5, 50)

y1 = np.sqrt(x)
y2 = x**3
y3 = 2**x
y4 = 1/(x+1)


plt.figure(figsize=(8,6))

plt.plot(x, y1, color='aqua', label='$y=\sqrt{x}$')

plt.plot(x, y2, color='indigo', label='$y=x^3$')

plt.plot(x, y3, color='mediumaquamarine', label='$y=2^x$')

plt.plot(x, y4, color='forestgreen', label= '$y=1/(x+1)$')

plt.legend(loc='upper center')

plt.xlabel('X')
plt.ylabel('Y')

plt.grid(ls='-.')

plt.show
```
